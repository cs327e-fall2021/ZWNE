```python
%%timeit
import mysql.connector

connection = mysql.connector.connect(
  host="10.128.0.4",
  user="root",
  password="zwne",
  database="load_testing",
  autocommit=False
)

print(connection)

df = pd.read_csv('us-1000000.csv') 
rows = df.values.tolist()
total = 0
count = 0
batch_size = 5000
records = []

try:
    for row in rows:
        sql = "INSERT INTO Person (first_name, last_name, company, address, city, county, state_prov, zip, ph1, ph2, email, web)\
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        rec = row
        records.append(row)
        total += 1
        count += 1
        
        if count == batch_size:  
            cursor = connection.cursor()
            cursor.executemany(sql, records)
            connection.commit()
            print(total, "records inserted successfully into Person table")
            count = 0
            records = []
            cursor.close()
    
except mysql.connector.Error as error:
        print("Failed to insert record into Person table {}".format(error))

finally:
    if connection.is_connected():
        connection.close()
        print("MySQL connection is closed")
```
