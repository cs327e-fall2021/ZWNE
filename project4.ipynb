##### 2) create and populate BQ tables:

```python
import os
```

```python
gsutil_cmd = "gsutil ls gs://cs327e-open-access/school-enrollments/*"
file_listings = os.popen(gsutil_cmd)

for file in file_listings:

    start_index = file.rindex("_") + 1
    end_index = file.rindex(".")
    table = file[start_index:end_index]
    
    if table in ("co", "district", "fl", "ga", "id", "ma", "mo", "ne", "ny", "or", "school", "sd", "wi"):
        print("skipping " + table)
        continue

    bq_cmd = "bq --location=US load --autodetect --skip_leading_rows=1 "\
             "--source_format=CSV " + dataset_id + "." + table + ' ' + file
    print(bq_cmd)
    
    os.system(bq_cmd)
```

#### 3) Find school enrollments by school name, grade level, and year for Texas 

```python
!bq show --schema=true {dataset_id}.tx
```

#### Note: STRING type for "total" column

```python
%%bigquery
select school, grade, year, sum(cast(replace(replace(total, ",", ""), "<", "") as int64)) as total
from `platinum-analog-324601.school_enrollments.tx`
where total != "" and school != "" and district like '%AUSTIN%'
group by school, grade, year
order by school, grade, year
```

##### Note: BQ string functions documentation: 
##### https://cloud.google.com/bigquery/docs/reference/standard-sql/string_functions

```python
!bq --location=US mk --dataset "views"
```

```python
%%bigquery
create view views.austin_enrollments_by_school_grade_year as
select school, grade, year, sum(cast(replace(replace(total, ",", ""), "<", "") as int64)) as total
from `platinum-analog-324601.school_enrollments.tx`
where total != "" and school != "" and district like '%AUSTIN%'
group by school, grade, year
order by school, grade, year
```

```python
%%bigquery
select * from views.austin_enrollments_by_school_grade_year
```

```python
#Query 1: list all the Anchorage Schools with black students and the number of black students in each grade
```

```python
%%bigquery
select school,sum(cast(black as int)),grade
from `platinum-analog-324601.school_enrollments.ak`
where school != ""
and black !=0
and district = "Anchorage Schools"
group by school, black,grade
order by school,grade
```

```python
#Query 2: list enrollment data of schools in Charleston 01 district of SC
```

```python
%%bigquery
select district,year,grade,school, sum(total) as chasc_enroll from `platinum-analog-324601.school_enrollments.sc`
where district='Charleston 01' and school !=""
group by district,total,year,grade,school
having chasc_enroll>0
order by year,grade
```

```python
#Query 3: looking at changes in unhoused enrollment across WA districts by year
```

```python
%%bigquery
SELECT district, sum(homeless) as total_unhoused, year FROM `platinum-analog-324601.school_enrollments.wa` 

group by district, year 

having sum(homeless)>0 

order by district, year
```

```python
#Query 4: Total enrollment by grade in NV for 2019
```

```python
%%bigquery
select year, grade, sum(total) as enroll_by_grade from `platinum-analog-324601.school_enrollments.nv` 

where grade != "pk_12_total" and grade != "k_12_total" and year=2019

group by year, grade

order by grade 
```

```python
#Query 5: Percentage of district enrollment that is Hispanic by year in Wyoming
```

```python
%%bigquery
select year, district, sum(hispanic) as  total_hispanic, sum(total) as total_enroll, (sum(cast(hispanic as int))/sum(total)) as pct_hispanic from `platinum-analog-324601.school_enrollments.wy` 

group by year, district

having total_hispanic > 0 

order by district
```

```python
#Query 6: A list of schools in Anchorage Schools district of AK
```

```python
%%bigquery
select district, count(school) as schools_in_dist
from platinum-analog-324601.school_enrollments.ak
where school != "" and year=2020
group by district
order by district
```

```python
#Query 7: District enrollment by ethnicity for Wyoming, Washington school districts 
```

```python
%%bigquery
select state, district, sum(hispanic) as  total_hispanic, sum(black) as  total_black, sum(asian) as  total_asian, 
sum(white) as  total_white, sum(native_american_alaskan_native) as total_na_an, 
sum(native_hawaiian_pacific_islander) as total_nh_pi, sum(multiracial) as total_multiracial from `platinum-analog-324601.school_enrollments.wy`
group by state, district 
union all 
select state, district, sum(hispanic) as  total_hispanic, sum(black) as  total_black, sum(asian) as  total_asian, 
sum(white) as  total_white, sum(native_american_alaskan_native) as total_na_an, 
sum(native_hawaiian_pacific_islander) as total_nh_pi, sum(multiracial) as total_multiracial from `platinum-analog-324601.school_enrollments.wa`
group by state, district
order by state, district
```

```python
#Query 8: Male enrollment of WY and NV by district
```

```python
%%bigquery
select state, district,sum(male)as total_male from `platinum-analog-324601.school_enrollments.wy`
group by state, district
union all
select state, district, sum(male) as total_male from `platinum-analog-324601.school_enrollments.nv`
group by state, district
having total_male > 0 
order by state, district
```
