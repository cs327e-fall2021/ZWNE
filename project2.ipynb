```python
# inner join 1: list US customers
!psql -c "create table pagila.US_customers as select customer.last_name, customer.first_name, country.country from \
pagila.customer join pagila.address on customer.address_id=address.address_id join pagila.city on address.city_id=city.city_id \
join pagila.country on city.country_id=country.country_id where country.country_id=103 order by last_name"
```

```python
!psql -c "select * from pagila.US_customers limit 10"
```

```python
# inner join 2: list Action films
!psql -c "create table pagila.action_films as select film.title, category.name from pagila.film join pagila.film_category on film.film_id=film_category.film_id \
join pagila.category on film_category.category_id=category.category_id where category.category_id=1 order by film.title"
```

```python
!psql -c "select * from pagila.action_films limit 10"
```

```python
# inner join 3: list stores in Alberta
!psql -c "create table pagila.alberta_stores as select store.store_id, city.city from pagila.store join pagila.address \
on store.address_id=address.address_id join pagila.city on address.city_id=city.city_id where city.city_id=300 order by store_id"
```

```python
!psql -c "select * from pagila.alberta_stores"
```

```python
# outer join 1: list cities in China including city and country
!psql -c "create table pagila.cities_in_china as select city.city, country.country from pagila.city full join pagila.country on \
city.country_id=country.country_id where city.country_id=23 order by city.city"
```

```python
!psql -c "select * from pagila.cities_in_china limit 10"
```

```python
# outer join 2: left join film info with language, will not list languages with no associated films
!psql -c "create table pagila.film_with_language as select film.title, language.name from pagila.film left join pagila.language on \
film.language_id = language.language_id where film.rating='G' order by film.title"
```

```python
!psql -c "select * from pagila.film_with_language limit 10"
```

```python
# outer join 3: list rentals of Academy Dinosaur including rental date and return date, if present
!psql -c "create table pagila.rental_list as select film.title, rental.rental_date, rental.return_date from pagila.film join pagila.inventory \
on film.film_id=inventory.film_id full join pagila.rental on inventory.inventory_id=rental.inventory_id where film.film_id=1 order by rental.rental_date"
```

```python
!psql -c "select * from pagila.rental_list limit 10"
```

```python
#self-join: films with same category
!psql -c "create table pagila.same_category as select a.title as titlea, b.title as titleb, b.category from pagila.film_list a join \
pagila.film_list b on a.fid!=b.fid and a.category=b.category"
```

```python
!psql -c "select * from pagila.same_category limit 10"
```
